```python
import pandas as pd
import numpy as np
import re
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
print(os.listdir(nltk.data.find("corpora")))
```

```python
from nltk.corpus import movie_reviews
```

```python
movie_reviews.categories()
```

```python
pos_rev=movie_reviews.fileids('pos')
neg_rev=movie_reviews.fileids('neg')
```

```python
print(movie_reviews.fileids('pos')[:10])
print(len(movie_reviews.fileids('pos')))
```

```python
print(movie_reviews.fileids('neg')[:10])
print(len(movie_reviews.fileids('neg')))
```

```python
rev = movie_reviews.words('neg/cv000_29416.txt')
rev
```

```python
rev_list=list()
for rev in neg_rev:
    rev=movie_reviews.words(rev)
    rev_single=' '.join(rev)
    #re.sub('\W+', '', rev_single)
    rev_single=rev_single.replace(' .', '.')
    rev_single=rev_single.replace(' ,', ',')
    rev_single=rev_single.replace(' :', ':')
    rev_single=rev_single.replace("\' ", "'")
    rev_single=rev_single.replace(" \'", "'")
    rev_single=rev_single.replace(' " ', ' "')
    rev_single=rev_single.replace(' ?', '?')
    rev_single=rev_single.replace(' )', ')')
    rev_single=rev_single.replace('( ', '(')
    rev_list.append(rev_single)
```

```python
rev_list[:10]
```

```python
for rev in pos_rev:
    rev=movie_reviews.words(rev)
    rev_single=' '.join(rev)
    #re.sub('\W+', '', rev_single)
    rev_single=rev_single.replace(' .', '.')
    rev_single=rev_single.replace(' ,', ',')
    rev_single=rev_single.replace(' :', ':')
    rev_single=rev_single.replace("\' ", "'")
    rev_single=rev_single.replace(" \'", "'")
    rev_single=rev_single.replace(' " ', ' "')
    rev_single=rev_single.replace(' ?', '?')
    rev_single=rev_single.replace(' )', ')')
    rev_single=rev_single.replace('( ', '(')
    rev_list.append(rev_single)
```

```python
len(rev_list)
```

```python
neg_targets=np.zeros((1000,), dtype=np.int)
pos_targets=np.ones((1000,), dtype=np.int)
```

```python
target_list=[]
for target in neg_targets:
    target_list.append(target)
for target in pos_targets:
    target_list.append(target)
```

```python
len(target_list)
```

```python
y=pd.Series(target_list)
```

```python
count_vector=CountVectorizer(lowercase=True, stop_words='english', min_df=2)
```

```python
X_count_vector=count_vector.fit_transform(rev_list)
```

```python
X_count_vector.shape
```

```python
X_names=count_vector.get_feature_names()
X_names[:10]
```

```python
X_count_vector=pd.DataFrame(X_count_vector.toarray(), columns=X_names)
```

```python
X_count_vector.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_count_vector, y, test_size=0.25, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gauss_coff = GaussianNB()
g_predict=gauss_coff.fit(X_train, y_train).predict(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mult_coeff=MultinomialNB()
```

```python
m_predict = mult_coeff.fit(X_train, y_train).predict(X_test)
```

```python
print(confusion_matrix(y_test, g_predict))
```

```python
print(metrics.accuracy_score(y_test, g_predict))
```

```python
print(confusion_matrix(y_test, m_predict))
```

```python
print(metrics.accuracy_score(y_test, m_predict))
```
